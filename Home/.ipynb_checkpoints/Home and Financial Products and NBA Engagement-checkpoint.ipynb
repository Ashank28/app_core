### Creating temporary 1 week base for app login which will be used in all codes:-

```sql
create or replace table  `pypl-edw`.pp_scratch.temp_app_login_base_US as
select
         a.cust_id
       ,date_trunc(sessn_start_dt,week) as login_week_dt
       ,a.sessn_start_dt
       ,a.sessn_key
from `pypl-edw`.pp_product_analytics_datastore_views.cxp_login_active a
where sessn_start_dt between '2025-04-13' and '2025-04-19'
and rtrim(a.channel) = 'APP'
and fin_acct_type_code in ('PERSONAL','PREMIER') --New segmentation
--  and cust_acct_clsfn_key in (1,7) --older logic for consumer filter
and lower(client_os_catg) in ('android','ios')
and trim(prmry_reside_cntry_code) = 'US' --update this filter for DE
group by all;
```

```sql
select login_week_dt, count(distinct cust_id) as custs, count(distinct sessn_key) as sessns
from `pypl-edw`.pp_scratch.temp_app_login_base_US
group by 1;
```

## 1. Engagement on Homepage - Module level events
#### Note: At module level we can only track a module's available counts, impressions and scroll - clicks are present at tile level

```sql
select 
         date_trunc(a.sessn_start_dt,week) as week_dt
        ,a.module
        --Module level events:
        ,count(distinct case when event_name='dw_dashboard_module_loaded' then a.cust_id end) as module_custs_av
        ,count(distinct case when event_name='dw_dashboard_module_shown' then a.cust_id end) as module_custs_im
        ,count(distinct case when event_name='dw_dashboard_module_scrolled' then a.cust_id end) as module_custs_scrolled

        --Tile level events:
        ,count(distinct case when event_name='dw_dashboard_tile_loaded' then a.cust_id end) as tile_custs_av
        ,count(distinct case when event_name='dw_dashboard_tile_shown' then a.cust_id end) as tile_custs_im
        ,count(distinct case when event_name='dw_dashboard_tile_pressed' then a.cust_id end) as tile_custs_clicked

from `pypl-edw`.pp_peak_views.dw_ql_time_position_ssot a --Homepage events peak table
join `pypl-edw`.pp_scratch.temp_app_login_base_US as b --Restricting for US app logins
on a.cust_id=b.cust_id
and a.sessn_start_dt=b.sessn_start_dt

where a.channel='APP'
and a.sessn_start_dt between '2025-04-13' and '2025-04-19'
-- and a.cust_acct_clsfn_key in (1,7)
group by 1,2
order by 1,3 desc;
```

## 2. Engagement on Homepage - Tile level events
#### Note: At tile level, looking at the breakdown of 1)Financial product module and 2) activity feed makes sense; rest of the modules don't have finite cards

```sql
select 
         date_trunc(a.sessn_start_dt,week) as week_dt
        ,a.module
        ,a.tile

        --Tile level events:
        ,count(distinct case when event_name='dw_dashboard_tile_loaded' then a.cust_id end) as tile_custs_av
        ,count(distinct case when event_name='dw_dashboard_tile_shown' then a.cust_id end) as tile_custs_im
        ,count(distinct case when event_name='dw_dashboard_tile_pressed' then a.cust_id end) as tile_custs_clicked

from `pypl-edw`.pp_peak_views.dw_ql_time_position_ssot a --Homepage events peak table
join `pypl-edw`.pp_scratch.temp_app_login_base_US as b --Restricting for US app logins
on a.cust_id=b.cust_id
and a.sessn_start_dt=b.sessn_start_dt

where a.channel='APP'
and a.sessn_start_dt between '2025-04-13' and '2025-04-19'
and module in ('financial_products','activity_feed')
and tile not in ('{{card_type}}')
group by 1,2,3
order by 1,2,4 desc;
```

### 3. Breaking Financial Products module tiles further by state - Zero vs Enrolled vs Pending state

```sql
select 
         date_trunc(a.sessn_start_dt,week) as week_dt
        ,a.header_state
        ,a.tile

        --Tile level events:
        ,count(distinct case when event_name='dw_dashboard_tile_loaded' then a.cust_id end) as tile_custs_av
        ,count(distinct case when event_name='dw_dashboard_tile_shown' then a.cust_id end) as tile_custs_im
        ,count(distinct case when event_name='dw_dashboard_tile_pressed' then a.cust_id end) as tile_custs_clicked_overall --click on header or NBA
        
        ,count(distinct case when (link_name='header' or link_name is null) and event_name='dw_dashboard_tile_pressed' then a.cust_id end) as header_custs_clicked --click on header 
        ,count(distinct case when link_name='nba' and event_name='dw_dashboard_tile_pressed' then a.cust_id end) as nba_custs_clicked --click on NBA


from `pypl-edw`.pp_peak_views.dw_ql_time_position_ssot a --Homepage events peak table
join `pypl-edw`.pp_scratch.temp_app_login_base_US as b --Restricting for US app logins
on a.cust_id=b.cust_id
and a.sessn_start_dt=b.sessn_start_dt

where a.channel='APP'
and a.sessn_start_dt between '2025-04-13' and '2025-04-19'
and module in ('financial_products')
and header_state is not null

group by 1,2,3
order by 1,2,4 desc;
```

### 4. Engagement on each NBA (next best action) for each product within FS:

```sql
select 
         date_trunc(a.sessn_start_dt,week) as week_dt
        ,a.tile
        ,next_best_action
        --Tile level events:
        ,count(distinct case when event_name='dw_dashboard_tile_loaded' then a.cust_id end) as tile_custs_av
        ,count(distinct case when event_name='dw_dashboard_tile_shown' then a.cust_id end) as tile_custs_im
        ,count(distinct case when event_name='dw_dashboard_tile_pressed' then a.cust_id end) as tile_custs_clicked_overall --click on header or NBA
        
        ,count(distinct case when (link_name='header' or link_name is null) and event_name='dw_dashboard_tile_pressed' then a.cust_id end) as header_custs_clicked --click on header 
        ,count(distinct case when link_name='nba' and event_name='dw_dashboard_tile_pressed' then a.cust_id end) as nba_custs_clicked --click on NBA


from `pypl-edw`.pp_peak_views.dw_ql_time_position_ssot a --Homepage events peak table
join `pypl-edw`.pp_scratch.temp_app_login_base_US as b --Restricting for US app logins
on a.cust_id=b.cust_id
and a.sessn_start_dt=b.sessn_start_dt

where a.channel='APP'
and a.sessn_start_dt between '2025-04-13' and '2025-04-19'
and module in ('financial_products')
and header_state = 'enrolled'

group by 1,2,3
order by 1,2,4 desc;
```
