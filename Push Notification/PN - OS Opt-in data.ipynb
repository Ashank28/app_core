```sql
--Getting info about previous week which needs to be now added to the dataset

delete from pypl-edw.pp_oap_sing_hawk_t.push_notification_OS_optin_status WHERE evnt_dt between current_date-4 and current_date-1;
insert into pypl-edw.pp_oap_sing_hawk_t.push_notification_OS_optin_status 
select 
    evnt_dt, 
    evnt_ts,
    case 
        WHEN
             (upper(a.client_os) like 'ANDROID' or upper(a.client_os) like '%LG%' or upper(a.client_os) like '%SONY%' or  
              upper(a.client_os) like '%SAMSUNG%' or upper(a.client_os) like '%L4%' or upper(a.client_os) like '%NOKIA%' or 
              upper(a.client_os) like '%AND%' or upper(a.client_os) like 'RIM' or 
              upper(a.client_os) like '%LINUX%' or upper(a.client_os) like '%WIN%' or upper(a.client_os) like 'WI' )
             THEN 'ANDROID'
        WHEN 
            (upper(a.client_os) like 'IOS' or upper(a.client_os) like 'OS X' or upper(a.client_os) like 'MAC OS' or upper(a.client_os) like '%OS%') 
            THEN 'IOS'
        ELSE 'IOS'
    END AS app_platform,
    
    client_os,
    
    lower(substr(`pypl-edw`.cw_udf.cw_td_nvp(a.payload, 'opt_sel', '&', '=', 1), 1, 100)) as option_selected, --PN status: allowed/notdecided/declined
    
    a.cust_id, 
    sessn_key
    
from pypl-edw.pp_polestar_views.client_side_events a 
WHERE
    a.evnt_dt between current_date-4 and current_date-1
    and lower(a.event_name) in 
    (
        'app_push_notification_opt_in_status_fetched' --when someone logs into the app, this event is triggered which captures user's PN opt-in status
    )

group by 1, 2, 3, 4, 5, 6, 7
;
```

```sql
--Taking only latest OS Opt-in status of users

create or replace table pypl-edw.pp_scratch.push_notification_OS_optin_status_latest_status
cluster by evnt_dt, cust_id as 

select 
    evnt_dt,
    evnt_ts,
    app_platform, 
    client_os,
    cust_id,
    option_selected
from
(

    select 
        evnt_dt,
        evnt_ts,
        app_platform,
        client_os,
        cust_id,
        option_selected,
        ROW_NUMBER() over (Partition by cust_id, app_platform order by evnt_ts desc) as rn

    from pypl-edw.pp_oap_sing_hawk_t.push_notification_OS_optin_status
) a

where a.rn = 1;
```
